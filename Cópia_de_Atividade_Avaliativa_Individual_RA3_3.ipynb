# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 3, 1000)
y1 = x * np.exp(-x**2)
y2 = x + 1
y3 = np.full_like(x, 2)

a = np.min(x)
b = np.max(x)

area = np.trapz(np.abs(y1 - y2), x)

plt.plot(x, y1, label='y = xe^(-x^2)')
plt.plot(x, y2, label='y = x + 1')
plt.plot(x, y3, label='x = 2')
plt.fill_between(x, y1, y2, where=(y1 >= y2), facecolor='blue', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

print('Área entre as curvas:', area)
```

Resolva, algebricamente, explicitando em latex todos os passos necessários, as seguintes integrais:

##a)

a) 

Podemos utilizar a fórumula da integração por partes:

$
\int u \, dv = uv - \int v \, du
$

Vamos considerar $u = \cos(3t)$ e $dv = \sin(8t) \, dt$. Dessa forma, podemos determinar $du$ e $v$ respectivamente como:

$
du = -3\sin(3t) \, dt \quad \text{e} \quad v = -\frac{1}{8}\cos(8t)
$

Aplicando a fórmula de integração por partes:

$
\int \cos(3t) \sin(8t) \, dt = -\frac{1}{8}\cos(8t)\cos(3t) - \int \left(-\frac{1}{8}\cos(8t)\right)\left(-3\sin(3t)\right) \, dt
$

Simplificando, temos:

$
\int \cos(3t) \sin(8t) \, dt = -\frac{1}{8}\cos(8t)\cos(3t) + \frac{3}{8}\int \cos(8t)\sin(3t) \, dt
$

Resolvendo integral restante $\int \cos(8t)\sin(3t) \, dt$ utilizando novamente a fórmula de integração por partes. Vamos considerar $u = \cos(8t)$ e $dv = \sin(3t) \, dt$. Dessa forma, podemos determinar $du$ e $v$ respectivamente como:

$
du = -8\sin(8t) \, dt \quad \text{e} \quad v = -\frac{1}{3}\cos(3t)
$

Aplicando a fórmula de integração por partes, obtemos:

$
\int \cos(8t)\sin(3t) \, dt = -\frac{1}{3}\cos(8t)\cos(3t) - \int \left(-\frac{1}{3}\cos(8t)\right)\left(-8\sin(3t)\right) \, dt
$

Simplificando, temos:

$
\int \cos(8t)\sin(3t) \, dt = -\frac{1}{3}\cos(8t)\cos(3t) + \frac{8}{3}\int \cos(8t)\sin(3t) \, dt
$

Agora, vamos isolar a integral restante e resolver a equação:

$
\int \cos(8t)\sin(3t) \, dt = -\frac{1}{3}\cos(8t)\cos(3t) + \frac{8}{3}\int \cos(8t)\sin(3t) \, dt
$

$
\left(1 - \frac{8}{3}\right) \int \cos(8t)\sin(3t) \, dt = -\frac{1}{3}\cos(8t)\cos(3t)
$

$
-\frac{5}{3} \int \cos(8t)\sin(3t) \, dt = -\frac{1}{3}\cos(8t)\cos(3t)
$

Dividindo ambos os lados por $-\frac{5}{3}$, temos:

$
\int \cos(8t)\sin(3t) \, dt = \frac{1}{5}\cos(8t)\cos(3t)
$

Agora, substituímos esse resultado na integral original:

$
\int \cos(3t) \sin(8t) \, dt = -\frac{1}{8}\cos(8t)\cos(3t) + \frac{3}{8}\left(\frac{1}{5}\cos(8t)\cos(3t)\right) + C
$

Simplificando a expressão final, temos:

$
\int \cos(3t) \sin(8t) \, dt = \frac{1}{40}\cos(8t)\cos(3t) + C
$

Portanto, a solução da integral $\int \cos(3t) \sin(8t) \, dt$ é $\frac{1}{40}\cos(8t)\cos(3t) + C$.

$
\int 4x \cos(2 - 3x) \, dx = \int 4x \cos(u) \left(-\frac{1}{3} \, du\right) = -\frac{4}{3} \int x \cos(u) \, du
$

Agora, precisamos encontrar $x$ em termos de $u$ para substituir na integral. Podemos fazer isso resolvendo a equação $u = 2 - 3x$ em relação a $x$:

$
3x = 2 - u \implies x = \frac{2 - u}{3}
$

Substituindo $x$ em termos de $u$ na integral, temos:

$
-\frac{4}{3} \int \left(\frac{2 - u}{3}\right) \cos(u) \, du
$

Simplificando temos:

$
-\frac{4}{3} \int \left(\frac{2}{3} - \frac{u}{3}\right) \cos(u) \, du
$

Expandindo a expressão dentro da integral, temos:

$
-\frac{4}{3} \int \left(\frac{2}{3} \cos(u) - \frac{u}{3} \cos(u)\right) \, du
$

Podemos integrar termo-a-termo:

$
-\frac{4}{3} \left(\frac{2}{3} \int \cos(u) \, du - \frac{1}{3} \int u \cos(u) \, du\right)
$

Integrando cada termo separadamente, temos:

$
-\frac{4}{3} \left(\frac{2}{3} \sin(u) - \frac{1}{3} \int u \cos(u) \, du\right)
$

Agora, precisamos resolver a integral $\int u \cos(u) \, du$. Para isso, utilizaremos a integração por partes, considerando $u = u$ e $dv = \cos(u) \, du$. Aplicamos a fórmula de integração por partes, assim, temos:

$

Integrando $\int \sin(u) \, du$, temos $-\cos(u)$. Substituindo na integral original, temos:

$
-\frac{4}{3} \left(\frac{2}{3} \sin(u) - \frac{1}{3} \left(u \sin(u) - (-\cos(u))\right)\right)
$

Simplificando a expressão, temos:

$

Agora, substituímos $u = 2 - 3x$ de volta na expressão:

$

Simplificando ainda mais, temos:

$
-\frac{8}{9} \sin(2 - 3x) + \frac{4}{9} (2 - 3x) \sin(2 - 3x) - \frac{4}{9} \cos(2 - 3x)
$

Portanto, a solução da integral $\int 4x \cos(2 - 3x) \, dx$ é $-\frac{8}{9} \sin(2 - 3x) + \frac{4}{9} (2 - 3x) \sin(2 - 3x) - \frac{4}{9} \cos(2 - 3x)$.


# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração. 

##a)
$$\int_0^2 x^2+1 \space dx$$

##b) 
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



```python
#a)
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 1

a = 0
b = 2

x = np.linspace(-1, 3, 100)
y = f(x)

integral = np.trapz(y, x)

def F(x):
    return (1/3) * x**3 + x

x_integral = np.linspace(-1, 3, 100)
y_integral = F(x_integral)

plt.plot(x, y, label='Função original: $x^2 + 1$')
plt.plot(x_integral, y_integral, label='Função após integração: $\\frac{1}{3}x^3 + x$')
plt.fill_between(x_integral, y_integral, alpha=0.2)

plt.axvline(x=a, color='r', linestyle='--', label='Limite inferior')
plt.axvline(x=b, color='g', linestyle='--', label='Limite superior')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função')
plt.legend()

plt.show()

print('Valor da integral:', integral)
```
