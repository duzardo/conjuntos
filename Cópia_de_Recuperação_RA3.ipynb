#Questão 1 (vale até 2 pontos)
Determine a área entre as curvas definidas por $x_1=y^2-2$ e $x_2=y$. Trace o gráfico destas funções sombreando em azul claro a área entre elas. Você pode usar apenas o python, matplotlib, numpy e sympy. Neste caderno do Google Colab.

## Resposta

```python
from sympy import symbols, solve, integrate, N
import numpy as np
import matplotlib.pyplot as plt

y = symbols('y')
x1 = y**2 - 2
x2 = y

intersections = solve(x1 - x2, y)

area = N(integrate(abs(x1 - x2), (y, intersections[0], intersections[1])))
print("A área entre as curvas é: ", area)

y_values = np.linspace(float(intersections[0]), float(intersections[1]), 400)
x1_values = y_values**2 - 2
x2_values = y_values

plt.figure(figsize=(6, 6))
plt.plot(x1_values, y_values, label='x1 = y² - 2')
plt.plot(x2_values, y_values, label='x2 = y')
plt.fill_betweenx(y_values, x1_values, x2_values, color='lightblue', label='Área entre as curvas')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Área entre as curvas x1 = y² - 2 e x2 = y')
plt.grid(True)
plt.show()
```

 Para encontrar a área as duas funções, podemos subtrair a integral da função inferior da integral da função superior. A área A entre as funções $y1 = e^x$, $y2 = x² - 1$ nos limites $x1 = -1$ e $x2 = 1$ é dada por:

 $A = \int_{-1}^{1} (y1 - y2) dx = \int_{-1}^{1} (e^x - (x^2 - 1)) dx$

 Essa integral pode ser separada em duas partes:

 $A = \int_{-1}^{1} e^x dx - \int_{-1}^{1} x^2 dx + \int_{-1}^{1} dx$

 Podemos resolver cada uma separadamente.
 A integral de e^x é a própria função e^x, então:

 $\int_{-1}^{1} e^x dx = [e^x]_{-1}^{1} = e - \frac{1}{e}$

 A integral de x^2 é (1/3)x³, então:

 $\int_{-1}^{1} x^2 dx = [\frac{1}{3}x^3]_{-1}^{1} = \frac{1}{3} - \frac{1}{3} = 0$

 A integral de 1 é x, então:

 $\int_{-1}^{1} dx = [x]_{-1}^{1} = 1 - (-1) = 2
$

Substituindo os valores de volta na expressão original, temos:

$A = (e - \frac{1}{e}) - 0 + 2 = e + 2 - \frac{1}{e}$

Portanto, a área entre as funções $y1 = e^x$ e $y2 = x² - 1$ nos limites $x = -1$ e $x = 1$ é $e + 2 - 1/e$

Um objeto está se movendo e experimenta diferentes acelerações em diferentes tempos. Os tempos e as respectivas acelerações são:

1. No instante $t = 1$ segundo, aceleração $a = 4 m/s²$
2. No instante $t = 3$ segundos, aceleração $a = -2 m/s²$
3. No instante $t = 4$ segundos, aceleração $a = 6 m/s²$
4. No instante $t = 6$ segundos, aceleração $a = 1 m/s²$
A posição inicial do objeto é $s_0 = 0 m$ e a velocidade inicial é $v_0 = 0 m/s$.

Você precisa imprimir com qualidade matemática (latex) a função encontrada para a aceleração, a função encontrada para a posição e a posição no instante $t=4.2s$. Além, é claro, dos gráficos de aceleração e posição.

Você poderá usar apenas as bibliotecas matplotlib, numpy e sympy.

## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt

t_values = np.array([1, 3, 4, 6])
a_values = np.array([4, -2, 6, 1])
t_eval = 4.2

def find_acceleration_function(t, t_values, a_values):
    a_interp = np.interp(t, t_values, a_values)
    return a_interp

def find_position_function(t, t_values, a_values):
    v_func = np.cumsum(find_acceleration_function(t, t_values, a_values))
    position_func = np.cumsum(v_func)
    return position_func

position_eval = find_position_function(t_eval, t_values, a_values)[-1]

print(f'Função de Aceleração: a(t)')
print(f'Função de Posição: s(t)')
print(f'Posição no instante t = {t_eval}s: {position_eval}m')

t = np.linspace(t_values[0], t_values[-1], 100)
a = find_acceleration_function(t, t_values, a_values)
s = find_position_function(t, t_values, a_values)

plt.figure()
plt.plot(t_values, a_values, 'bo-', label='Aceleração')
plt.xlabel('Tempo (s)')
plt.ylabel('Aceleração (m/s²)')
plt.title('Gráfico de Aceleração')
plt.grid(True)
plt.legend()

plt.figure()
plt.plot(t, s, 'r-', label='Posição')
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.title('Gráfico de Posição')
plt.grid(True)
plt.legend()

plt.show()
```

#Questão 4 (vale até 1 ponto)
Considere a função $f(x)$ dada por:

$$f(x) = 3x^2 - 4x + 2$$

Calcule a antiderivada $F(x)$ da função $f(x)$. Utilize o Teorema Fundamental do Cálculo para encontrar o valor exato da integral definida de $f(x)$ no intervalo $[1, 3]$.

Mostre todos os passos de cálculo, usando a notação matemática (latex) neste caderno do Google Colab e justifique suas respostas utilizando as propriedades das operações integração.

## Resposta



Primeiro, vamos encontrar a antiderivada da função. A antiderivada de uma função é a operação inversa da derivada, também conhecida como integral indefinida. Para a função $(f(x) = 3x^2 - 4x + 2)$, a antiderivada $(F(x))$ é dada por:

$[F(x) = \int f(x)$, $dx = \int (3x^2 - 4x + 2)$ , $dx]$

Separando os termos, temos:

$[F(x) = \int 3x^2 \, dx - \int 4x \, dx + \int 2 \, dx]$

Agora, usamos as regras básicas de integração. A integral de $(x^n)$ é $(\frac{1}{n+1}x^{n+1})$, e a integral de uma constante é a constante vezes \(x\). Então, temos:

$[F(x) = 3 \cdot \frac{1}{3}x^{2+1} - 4 \cdot \frac{1}{2}x^{1+1} + 2x + C]$

Simplificando, obtemos:

$[F(x) = x^3 - 2x^2 + 2x + C]$

A constante de integração $(C)$ surge porque a operação de derivada perde a informação sobre constantes.

Agora, vamos usar o Teorema Fundamental do Cálculo para encontrar a integral definida de $(f(x))$ no intervalo $([1, 3])$. O teorema nos diz que a integral definida de uma função $(f(x))$ entre os limites $(a)$ e $(b)$ é igual a $(F(b) - F(a))$, onde $(F(x))$ é a antiderivada de $(f(x))$.

Nesse caso, temos $(a = 1)$ e $(b = 3)$. Substituindo na fórmula, temos:

$[\int_{1}^{3} f(x) \, dx = F(3) - F(1)]$

Calculando $(F(3)) e (F(1))$, temos:

$[F(3) = (3)^3 - 2*(3)^2 + 2*3 = 27 - 18 + 6 = 15]$
$[F(1) = (1)^3 - 2*(1)^2 + 2*1 = 1 - 2 + 2 = 1]$

Substituindo $(F(3))$ e $(F(1))$ na fórmula, temos:

$[\int_{1}^{3} f(x) \, dx = 15 - 1 = 14]$

Portanto, a integral definida da função $(f(x))$ no intervalo $([1, 3])$ é 14.


#Questão 5 (vale até 2 pontos)
Considere as seguintes curvas no plano cartesiano:

Curva A: $y = x^2$
Curva B: $y = 2x - 1$

Encontre os valores de [a, b], de modo que a área entre estas curvas neste intervalo seja igual a 6 unidades quadradas sabendo que $a=2$.

Usando o Python e apenas as bibliotecas matplotlib, numpy e sympy,
1. crie um código que plote estas duas curvas no intervalo entre $-10 \leq x \leq 10$;
2. crie um código que econtre o valor desejado de $b$;
3. crie um código que plote estas duas curvas sombreando em azul claro a área entre elas limitada pelo intervalo $[a,b] de forma que seja possível ver esta área completamente;
4. prove que o valor de b que você encontrou produzirá uma área de 6 unidades quadradas com precisão maior ou igual a 0.001.


## Resposta

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

x = sp.Symbol('x')
curve_a = x**2
curve_b = 2*x - 1

a = 2
b = sp.Symbol('b')

x_vals = np.linspace(-10, 10, 1000)
y_vals_a = sp.lambdify(x, curve_a)(x_vals)
y_vals_b = sp.lambdify(x, curve_b)(x_vals)

plt.plot(x_vals, y_vals_a, label='Curva A: y = x^2')
plt.plot(x_vals, y_vals_b, label='Curva B: y = 2x - 1')

area_eq = sp.integrate(curve_a - curve_b, (x, a, b))
equation = sp.Eq(area_eq, 6)
b_solution = sp.solve(equation, b)

if len(b_solution) == 0:
    print("Não foi possível encontrar um valor de b que produza uma área de 6 unidades quadradas.")
else:
    b_value = b_solution[0]
    print("Valor de b:", b_value)
    x_fill = np.linspace(a, float(b_value), 100)
    y_fill_a = sp.lambdify(x, curve_a)(x_fill)
    y_fill_b = sp.lambdify(x, curve_b)(x_fill)
    plt.fill_between(x_fill, y_fill_a, y_fill_b, color='lightblue', alpha=0.5)

area_value = sp.integrate(curve_a - curve_b, (x, a, b_value)).evalf()
print("Área calculada:", area_value)

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curvas A e B')
plt.legend()
plt.grid(True)
plt.show()
```

b) A integral $$\int x \sin(x) \, dx$$ é resolvida usando a técnica de integração por partes.

Para a integral ∫ x sin(x) dx, escolhemos u = x e dv = sin(x) dx. Então, calculamos du = dx e v = -cos(x). Substituindo na fórmula original de integração por partes, obtemos:

\begin{align*}
\int x \sin(x) \, dx &= uv - \int v \, du \\
&= x(-\cos(x)) - \int -\cos(x) \, dx \\
&= -x \cos(x) + \int \cos(x) \, dx \\
&= -x \cos(x) + \sin(x) + C
\end{align*}
